(a) We use a Monte-Carlo integration method to generate code to estimate integral.

```python
N = int(input("Enter a large number N: ")) #prompt user to enter a large number.
a=0 #lower bound of integral.
b=2 # bound of integral.
V= (b-a)  # space restricting line in consideration.
sum_=0  # initiating sum_  before performing summation.
sum_1=0  # initiating sum_1 before performing summation.
for i in range(N):    # i from 0 to N. We repeat the following.
        x = rnd.uniform(a,b)    # picking x values randomly from range -r to r inside square.
        sum_+= (1/N)*(np.exp(-((x*x)/2))) # summing the function of integral over N range, i.e, f_bar.
        sum_1+= (1/N)*(np.exp(-((x*x)/2)))**2  # the function squared under the summution, i.e, f_bar_squared.   
u = V*sum_  + V*np.sqrt((sum_1 - (sum_)**2)/N)  # Monte-Carlo approximation subtracting error part, i.e, upper bound.  
l = V*sum_  - V*np.sqrt((sum_1 - (sum_)**2)/N)  # Monte-Carlo approximation subtracting error part, i.e, lower bound. 
print("Upper Estimated Integral =",u, "\t", "Lower Estimated Integral =", l) #print function to display upper and lower estimates.
```

(b) We use a Monte-Carlo integration method to generate code to calculate integral over a surface domain.

```python
n = int(input("Enter a large number N: ")) #prompt user to enter a large number.
V=4 # volume of surface domain.
sum_=0 # initaiting sum_
sum_1=0 # initiating sum_1
for i in range(n):         # i from 0 to n. 
        x = rnd.uniform(-1,1)   # picking x values randomly from range -r to r inside domain.
        y = rnd.uniform(-1,1)     # picking y values randomly from range -r to r inside domain.
        if (x**2 + y**2) < 1:    # condition to restrict points inside domain.
            sum_+= (1/n)*(np.exp(x*x + y*y))    # summing the function of integral over N range, i.e, f_bar.
            sum_1+= (1/n)*(np.exp(x*x + y*y))**2 # the function squared under the summution, i.e, f_bar_squared.  
U = V*sum_  + V*np.sqrt((sum_1 - (sum_)**2)/n) # Monte-Carlo approximation subtracting error part, i.e, upper bound. 
L = V*sum_  - V*np.sqrt((sum_1 - (sum_)**2)/n) # Monte-Carlo approximation subtracting error part, i.e, lower bound. 
print("Upper Estimated Integral =",U, "\t", "Lower Estimated Integral  =", L) #print function to display upper and lower estimates.
```
